## Solu 2 – Kirjastot ja data

> Aja tämä solu sellaisenaan – älä muuta mitään.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (9, 4)

data = {
    "pvm": [
        "2024-01-08", "2024-01-15", "2024-01-22",
        "2024-02-05", "2024-02-14", "2024-02-28",
        "2024-03-03", "2024-03-19", "2024-03-27",
        "2024-04-02", "2024-04-11", "2024-04-24",
        "2024-05-07", "2024-05-20", "2024-05-29",
        "2024-06-04", "2024-06-18", "2024-06-30"
    ],
    "tuote": [
        "Kannettava", "Hiiri", "Näyttö",
        "Hiiri", "Näppäimistö", "Kannettava",
        "Näyttö", "Kannettava", "Hiiri",
        "Näppäimistö", "Näyttö", "Kannettava",
        "Hiiri", "Kannettava", "Näppäimistö",
        "Näyttö", "Hiiri", "Kannettava"
    ],
    "kategoria": [
        "Tietokoneet", "Lisälaitteet", "Näytöt",
        "Lisälaitteet", "Lisälaitteet", "Tietokoneet",
        "Näytöt", "Tietokoneet", "Lisälaitteet",
        "Lisälaitteet", "Näytöt", "Tietokoneet",
        "Lisälaitteet", "Tietokoneet", "Lisälaitteet",
        "Näytöt", "Lisälaitteet", "Tietokoneet"
    ],
    "maara":  [2, 5, 1, 8, 4, 1, 2, 3, 6, 5, 1, 2, 10, 1, 3, 2, 7, 2],
    "hinta":  [899, 29, 349, 29, 59, 899, 349, 899, 29, 59, 349, 899, 29, 899, 59, 349, 29, 899],
    "alue": [
        "Pohjoinen", "Etelä", "Pohjoinen",
        "Länsi", "Etelä", "Pohjoinen",
        "Länsi", "Etelä", "Pohjoinen",
        "Länsi", "Etelä", "Pohjoinen",
        "Etelä", "Länsi", "Etelä",
        "Pohjoinen", "Länsi", "Etelä"
    ],
    "asiakastyyppi": [
        "Yritys", "Yksityinen", "Yritys",
        "Yksityinen", "Yritys", "Yritys",
        "Yksityinen", "Yritys", "Yksityinen",
        "Yritys", "Yksityinen", "Yritys",
        "Yksityinen", "Yritys", "Yksityinen",
        "Yritys", "Yksityinen", "Yritys"
    ]
}

df = pd.DataFrame(data)
df["pvm"]            = pd.to_datetime(df["pvm"])
df["kokonaismyynti"] = df["maara"] * df["hinta"]
df["kuukausi"]       = df["pvm"].dt.month

print("✅ Data ladattu onnistuneesti!")
print(f"   Rivejä: {len(df)}, Sarakkeita: {len(df.columns)}")
df.head(8)
```

## Solu 3 – Datan rakenne ja perustilastot

```python
print("=== DATAN RAKENNE ===")
df.info()
print()
print("=== PERUSTILASTOT ===")
df.describe()
```

## Solu 4 – Tuotteet, alueet ja asiakastyypit

```python
print("Tuotteet:")
print(df["tuote"].value_counts())
print()
print("Alueet:")
print(df["alue"].value_counts())
print()
print("Asiakastyypit:")
print(df["asiakastyyppi"].value_counts())
```

---
## Solu 5 – Myynti ryhmiteltynä

```python
myynti_tuote  = df.groupby("tuote")["kokonaismyynti"].sum().sort_values(ascending=False)
myynti_alue   = df.groupby("alue")["kokonaismyynti"].sum().sort_values(ascending=False)
myynti_kk     = df.groupby("kuukausi")["kokonaismyynti"].sum()
myynti_asiakas= df.groupby("asiakastyyppi")["kokonaismyynti"].sum()

print("Myynti tuotteittain (€):")
print(myynti_tuote)
print()
print("Myynti alueittain (€):")
print(myynti_alue)
print()
print("Kuukausittainen myynti (€):")
print(myynti_kk)
print()
print("Myynti asiakastyypeittäin (€):")
print(myynti_asiakas)
```

---
## Solu 6 – Kaavio 1: Myynti tuotteittain

```python
fig, ax = plt.subplots()
ax.bar(myynti_tuote.index, myynti_tuote.values,
       color="steelblue", edgecolor="white")
ax.set_title("Kokonaismyynti tuotteittain", fontsize=14, pad=12)
ax.set_xlabel("Tuote")
ax.set_ylabel("Myynti (€)")

for i, v in enumerate(myynti_tuote.values):
    ax.text(i, v + 30, f"{v:,} €", ha="center", fontsize=9)

plt.tight_layout()
plt.show()
```

---
## Solu 7 – Kaavio 2: Kuukausittainen myynti ja trendiviiva

**Trendiviiva** näyttää datan yleisen suunnan yksittäisten vaihtelujen takaa.  
NumPy sovittaa dataan suoran viivan `polyfit`-funktiolla.

```python
kuukaudet = myynti_kk.index.values
myynti    = myynti_kk.values

# Trendiviiva NumPy polyfit-funktiolla
z = np.polyfit(kuukaudet, myynti, 1)
p = np.poly1d(z)

fig, ax = plt.subplots()
ax.plot(kuukaudet, myynti, marker="o", color="teal",
        linewidth=2, markersize=7, label="Toteutunut myynti")
ax.plot(kuukaudet, p(kuukaudet), "--", color="red",
        linewidth=1.5, label="Trendi")

ax.set_title("Kuukausittainen myynti ja trendi 2024", fontsize=14, pad=12)
ax.set_xlabel("Kuukausi")
ax.set_ylabel("Myynti (€)")
ax.set_xticks(range(1, 7))
ax.set_xticklabels(["Tammi", "Helmi", "Maalis", "Huhti", "Touko", "Kesä"])
ax.legend()

plt.tight_layout()
plt.show()

# Tulkinta
if z[0] > 0:
    print(f"📈 Trendi on NOUSEVA – myynti kasvaa keskimäärin {z[0]:,.0f} €/kk")
else:
    print(f"📉 Trendi on LASKEVA – myynti laskee keskimäärin {abs(z[0]):,.0f} €/kk")
```

---
## Solu 8 – Kaavio 3: Korrelaatiomatriisi

**Korrelaatio** mittaa kahden muuttujan välistä yhteyttä.  
Arvo vaihtelee −1:stä +1:een: lähellä +1 = samaan suuntaan, lähellä −1 = vastakkaisiin suuntiin, 0 = ei yhteyttä.

```python
korrelaatio = df[["maara", "hinta", "kokonaismyynti"]].corr()

print("Korrelaatiomatriisi:")
print(korrelaatio.round(2))
print()

fig, ax = plt.subplots(figsize=(6, 4))
sns.heatmap(korrelaatio,
            annot=True,
            cmap="coolwarm",
            vmin=-1, vmax=1,
            fmt=".2f",
            ax=ax)
ax.set_title("Korrelaatiomatriisi: määrä, hinta, kokonaismyynti",
             fontsize=13, pad=12)
plt.tight_layout()
plt.show()

print("Tulkinta:")
print("  Lähellä +1.0 = muuttujat liikkuvat samaan suuntaan")
print("  Lähellä  0.0 = ei havaittavaa yhteyttä")
print("  Lähellä -1.0 = muuttujat liikkuvat vastakkaisiin suuntiin")
```

---
## Solu 9 – Pivot-taulukko: tuote × alue

**Pivot-taulukko** tiivistää datan kahden ulottuvuuden mukaan samanaikaisesti.  
Sen sijaan että katsotaan vain *mikä tuote myy eniten*, voidaan kysyä *mikä tuote myy eniten millä alueella?*

```python
pivot = pd.pivot_table(
    df,
    values   = "kokonaismyynti",
    index    = "tuote",
    columns  = "alue",
    aggfunc  = "sum",
    fill_value = 0
)

print("Kokonaismyynti (€) – tuote × alue:")
print(pivot)
print()
print(f"Paras tuote-alue-yhdistelmä: "
      f"{pivot.stack().idxmax()} → {pivot.stack().max():,} €")
```

---
## Solu 10 – Pivot visualisoituna heatmappina

```python
fig, ax = plt.subplots(figsize=(8, 4))
sns.heatmap(pivot,
            annot=True,
            fmt=",",
            cmap="YlGnBu",
            ax=ax)
ax.set_title("Myynti (€) tuotteittain ja alueittain", fontsize=13, pad=12)
plt.tight_layout()
plt.show()
```
